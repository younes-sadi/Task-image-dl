```python
import os
import numpy as np
from PIL import Image
import fiftyone as fo

dataset = fo.zoo.load_zoo_dataset(
    "open-images-v7",
    split="train",
    label_types=["segmentations"],
    classes=["Dog"],
    max_samples=1000,
    dataset_name="open-images-Dog",
    only_matching=True
)

dataset_2 = fo.zoo.load_zoo_dataset(
    "open-images-v7",
    split="train",
    label_types=["segmentations"],
    classes=["Bird"],
    max_samples=1000,
    dataset_name="open-images-bird",
    only_matching=True
)

dataset_3 = fo.zoo.load_zoo_dataset(
    "open-images-v7",
    split="train",
    label_types=["segmentations"],
    classes=["Cat"],
    max_samples=1000,
    dataset_name="open-images-cat",
    only_matching=True
)

dataset.merge_samples(dataset_2)
dataset.merge_samples(dataset_3)

root_dir = "dataset"
imgs_dir = os.path.join(root_dir, "imgs")
masks_dir = os.path.join(root_dir, "masks")
os.makedirs(imgs_dir, exist_ok=True)
os.makedirs(masks_dir, exist_ok=True)

data = []

for sample in dataset:
    img_src = sample.filepath
    img_name = os.path.basename(img_src)
    img_dst = os.path.join(imgs_dir, img_name)

    if not os.path.exists(img_dst):
        os.system(f'cp "{img_src}" "{img_dst}"')

    detections = sample["ground_truth"].detections

    for i, det in enumerate(detections):
        label = det.label
        mask = det.mask

        mask_filename = f"{os.path.splitext(img_name)[0]}_{i}.png"
        mask_path = os.path.join(masks_dir, mask_filename)

        mask_img = Image.fromarray(mask.astype(np.uint8) * 255)
        mask_img.save(mask_path)

        data.append({
            "img_path": f"images/{img_name}",
            "mask_path": f"masks/{mask_filename}",
            "label": label
        })
```

```python
import os
import cv2
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score, recall_score, f1_score
import fiftyone as fo

# ================== 1. Dataset Preparation ==================

# Download test images
test_dataset = fo.zoo.load_zoo_dataset(
    "open-images-v7",
    split="test",
    label_types=["segmentations"],
    classes=["Dog", "Cat", "Bird"],
    max_samples=100,
    dataset_name="open-images-test",
    only_matching=True
)

# Create directory structure
# ================== Configuration ==================
IMG_SIZE = 128  # Define this FIRST
NUM_CLASSES = 4  # Background + 3 classes
CLASS_NAMES = {1: "Dog", 2: "Cat", 3: "Bird"}

# ================== 1. Dataset Preparation ==================
def prepare_dataset(dataset, root_dir):
    os.makedirs(os.path.join(root_dir, "images"), exist_ok=True)
    os.makedirs(os.path.join(root_dir, "masks"), exist_ok=True)

    for sample in dataset:
        # Load original image
        img = cv2.imread(sample.filepath)
        h, w = img.shape[:2]

        # Create empty mask
        mask = np.zeros((h, w), dtype=np.uint8)

        # Process detections
        for det in sample["ground_truth"].detections:
            if det.mask is not None:
                # Resize detection mask to original image size
                resized_mask = cv2.resize(
                    det.mask.astype(np.uint8),
                    (w, h),
                    interpolation=cv2.INTER_NEAREST
                )
                class_id = 1 if det.label == "Dog" else 2 if det.label == "Cat" else 3
                mask = np.where(resized_mask, class_id, mask)

        # Resize to standard size
        img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
        mask = cv2.resize(mask, (IMG_SIZE, IMG_SIZE), interpolation=cv2.INTER_NEAREST)

        # Save files
        base_name = os.path.basename(sample.filepath)
        cv2.imwrite(os.path.join(root_dir, "images", base_name), img)
        cv2.imwrite(os.path.join(root_dir, "masks", base_name), mask)


# Prepare test set (100 images)
prepare_dataset(test_dataset, "/content/test_data")

# ================== 2. Data Loading & Preprocessing ==================
IMG_SIZE = 128
NUM_CLASSES = 4  # Background + 3 classes
BATCH_SIZE = 16

def load_data(data_dir):
    images = []
    masks = []

    valid_extensions = ['.jpg', '.jpeg', '.png']

    for img_file in os.listdir(os.path.join(data_dir, "images")):
        # Skip non-image files
        if not any(img_file.lower().endswith(ext) for ext in valid_extensions):
            continue

        # Load image
        img_path = os.path.join(data_dir, "imgs", img_file)
        img = cv2.imread(img_path)
        if img is None:
            print(f"Warning: Could not read image {img_path}")
            continue

        # Resize image
        img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
        img = img / 255.0

        # Load corresponding mask
        mask_path = os.path.join(data_dir, "masks", img_file)
        mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
        if mask is None:
            print(f"Warning: Could not read mask for {img_file}")
            continue

        # Process mask
        mask = cv2.resize(mask, (IMG_SIZE, IMG_SIZE), interpolation=cv2.INTER_NEAREST)
        mask = tf.keras.utils.to_categorical(mask, NUM_CLASSES)

        images.append(img)
        masks.append(mask)

    print(f"Successfully loaded {len(images)} images with masks")
    return np.array(images), np.array(masks)
# Load training data
X_train, y_train = load_data("/content/dataset")
X_test, y_test = load_data("/content/test_data")

```

```python
# ================== 3. Model Architecture ==================
def unet_model():
    inputs = layers.Input((IMG_SIZE, IMG_SIZE, 3))

    # Encoder
    c1 = layers.Conv2D(16, 3, activation='relu', padding='same')(inputs)
    c1 = layers.BatchNormalization()(c1)
    p1 = layers.MaxPooling2D()(c1)

    c2 = layers.Conv2D(32, 3, activation='relu', padding='same')(p1)
    c2 = layers.BatchNormalization()(c2)
    p2 = layers.MaxPooling2D()(c2)

    # Bottleneck
    c3 = layers.Conv2D(64, 3, activation='relu', padding='same')(p2)
    c3 = layers.BatchNormalization()(c3)

    # Decoder
    u1 = layers.UpSampling2D()(c3)
    u1 = layers.concatenate([u1, c2])
    c4 = layers.Conv2D(32, 3, activation='relu', padding='same')(u1)

    u2 = layers.UpSampling2D()(c4)
    u2 = layers.concatenate([u2, c1])
    c5 = layers.Conv2D(16, 3, activation='relu', padding='same')(u2)

    outputs = layers.Conv2D(NUM_CLASSES, 1, activation='softmax')(c5)

    return models.Model(inputs, outputs)

model = unet_model()

# ================== 4. Training Setup ==================
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy', tf.keras.metrics.MeanIoU(num_classes=NUM_CLASSES)]
)

callbacks = [
    tf.keras.callbacks.EarlyStopping(patience=5, restore_best_weights=True),
    tf.keras.callbacks.ModelCheckpoint('best_model.h5', save_best_only=True)
]

history = model.fit(
    X_train, y_train,
    validation_split=0.2,
    batch_size=BATCH_SIZE,
    epochs=30,
    callbacks=callbacks
)
```

```python

# ================== 5. Evaluation on 100 Test Images ==================
model.load_weights('best_model.h5')
y_pred = model.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=-1).flatten()
y_true_classes = np.argmax(y_test, axis=-1).flatten()

print("Metrics for 100 Unseen OpenImages Test Samples:")
for class_idx, class_name in enumerate(["Dog", "Cat", "Bird"]):
    precision = precision_score(y_true_classes == class_idx+1, y_pred_classes == class_idx+1, zero_division=0)
    recall = recall_score(y_true_classes == class_idx+1, y_pred_classes == class_idx+1, zero_division=0)
    f1 = f1_score(y_true_classes == class_idx+1, y_pred_classes == class_idx+1, zero_division=0)

    print(f"{class_name}:")
    print(f"  Precision: {precision:.3f}")
    print(f"  Recall: {recall:.3f}")
    print(f"  F1 Score: {f1:.3f}")
    print("-------------------")
```
